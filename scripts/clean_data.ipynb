# Clan salary

```python
def clean_salary(salary):
    if pd.isnull(salary) or salary in ['Not Disclosed', 'N/A', '']:
        return np.nan
    
    salary = str(salary).replace('000', 'K').replace(',', '')
    return float(salary)
df
```
